```
def group_plot(feat_groupby, feat_agg='price_value', agg_func=[np.mean, np.median, np.size], feat_sort='mean', top=50):
  return (
      df
      .groupby(feat_groupby)[feat_agg]
      .agg(agg_func)
      .sort_values(by=feat_sort, ascending=False)
      .head(top)
  ).plot(kind='bar', figsize=(15,5), subplots=True)

```

```
group_plot('param_kraj-pochodzenia', feat_sort='size');
```

```
group_plot('param_typ', feat_sort='mean');
```

```
group_plot('param_kolor', feat_sort='mean');
```
